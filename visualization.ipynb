# Visualization Practice in Python  


## **Seaborn**
provides high-level interface to `matplotlib`.    
Features:   
- aesthetically pleasing default themes  
- custom color palettes  
- visualizing information from matrices and DataFrames  
- easily and flexibly displaying distributions  
- making attractive stat plots  


```python
import numpy as np 
import pandas as pd 
import seaborn as sns  
sns.set(color_codes=True)  
import matplotlib.pyplot as plt  
import matplotlib as mpl  
import os  
```

```python
# load data
parks=pd.read_csv(os.path.join(data_path, "national_parks_species\\parks.csv"))
species=pd.read_csv(os.path.join(data_path, "national_parks_species\\species.csv"))
volcano=pd.read_csv(os.path.join(data_path, "volcanoes\\volcano.csv"))
```

### Line Charts

```python
plt.style.use('seaborn-darkgrid')

#Recover default matplotlib settings 
mpl.rcParams.update(mpl.rcParamsDefault)
%matplotlib inline
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(x="Latitude", y="Longitude", data=parks)
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(x="Latitude", y="Longitude", data=parks, linewidth=5)
plt.title("National Parks Longitude vs. Latitude")
```

```python
plt.figure(figsize=(20,6))
sns.lineplot(data=parks['Longitude'],linewidth = 1.5 , label = 'longitude')
sns.lineplot(data=parks['Latitude'],linewidth = 1.5 , label = 'latitude')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.set(rc={"axes.facecolor": "#283747", "axes.grid":False,
            'xtick.labelsize':10, 'ytick.labelsize':10 })
plt.title("Acres", fontsize=14)
plt.xticks(rotation=45) 
sns.lineplot(x =parks["Latitude"] , y =parks["Longitude"].index.values  , color = '#ff9900')
plt.show()
```

```python
# Show groups with different colors using "hue"
plt.figure(figsize=(14,7))
plt.style.use('seaborn-darkgrid')
plt.xticks(rotation=45) 

# Group variable using "hue" that will produce lines with different colors
sns.lineplot(x='primary_volcano_type' , y="elevation" , hue= 'tectonic_settings', data=volcano)
plt.show()
```

```python
# Show groups with different colors using "hue"
plt.figure(figsize=(14,7))
plt.style.use('seaborn-darkgrid')
plt.xticks(rotation=45) 

# Group variable using "hue" that will produce lines with different colors
sns.lineplot(x='primary_volcano_type' , y="elevation" , hue= 'tectonic_settings', style='tectonic_settings', markers=True, data=volcano)
plt.show()
```

```python
# Show groups with different colors using "hue"
plt.figure(figsize=(14,7))
plt.style.use('seaborn-darkgrid')
plt.xticks(rotation=45) 

# Group variable using "hue" that will produce lines with different colors
sns.lineplot(x='primary_volcano_type' , y="elevation" , hue= 'tectonic_settings',err_style="bars" , data=volcano)
plt.show()
```

```python
# Show groups with different colors using "hue"
plt.figure(figsize=(14,7))
plt.style.use('seaborn-darkgrid')
plt.xticks(rotation=45) 

# Group variable using "hue" that will produce lines with different colors
sns.lineplot(x='primary_volcano_type' , y="elevation" , hue= 'tectonic_settings',err_style="bars" , data=volcano)
sns.set(style='dark',)
plt.show()
```
